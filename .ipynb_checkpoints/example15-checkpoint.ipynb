In statistical data analysis the **TSS**

## <span class="section-number">6.5. </span>Statistical Tests<a href="#statistical-tests" class="headerlink"
title="Permalink to this headline"></a>

### <span class="section-number">6.5.1. </span>Correlational Test<a href="#correlational-test" class="headerlink"
title="Permalink to this headline"></a>

-   Pearson correlation: Tests for the strength of the association
    between two continuous variables.

-   Spearman correlation: Tests for the strength of the association
    between two ordinal variables (does not rely on the assumption of
    normal distributed data).

-   Chi-square: Tests for the strength of the association between two
    categorical variables.

### <span class="section-number">6.5.2. </span>Comparison of Means test<a href="#comparison-of-means-test" class="headerlink"
title="Permalink to this headline"></a>

-   Paired T-test: Tests for difference between two related variables.

-   Independent T-test: Tests for difference between two independent
    variables.

-   ANOVA: Tests the difference between group means after any other
    variance in the outcome variable is accounted for.

### <span class="section-number">6.5.3. </span>Non-parametric Test<a href="#non-parametric-test" class="headerlink"
title="Permalink to this headline"></a>

-   Wilcoxon rank-sum test: Tests for difference between two independent
    variables - takes into account magnitude and direction of
    difference.

-   Wilcoxon sign-rank test: Tests for difference between two related
    variables - takes into account magnitude and direction of
    difference.

-   Sign test: Tests if two related variables are different  ignores
    magnitude of change, only takes into account direction.
